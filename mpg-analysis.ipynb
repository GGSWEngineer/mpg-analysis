```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
mpg = sns.load_dataset("mpg")
```

### Task 2

Create a visual showing the relationship between horsepower and fuel economy.

```python
sns.scatterplot(x='horsepower', y='mpg', data=mpg)
plt.title('Relationship between Horsepower and Fuel Economy')
plt.xlabel('Horsepower')
plt.ylabel('Fuel Economy (mpg)')
plt.show()
```

 
The relationship between horsepower and fuel economy (mpg) reveals an inverse correlationâ€”lower horsepower is associated with higher fuel economy, while higher horsepower corresponds to lower fuel economy. This suggests that opting for a vehicle with lower horsepower may contribute to increased fuel efficiency and better mileage, making it a favorable choice for those prioritizing fuel economy. Conversely, higher horsepower vehicles tend to exhibit lower fuel efficiency, making them more suitable for individuals prioritizing performance over fuel economy.

### Task 3

Create a table summarizing fuel economy by country of origin

```python
summary_table = mpg.groupby("origin")["mpg"].agg(["count","mean","median","min","max"]).reset_index()
summary_table.columns = ["Country of Origin", "Number of Cars","Mean MPG","Median MPG", "Min MPG", "Max MPG"]
print(summary_table)
```

In the comparison of fuel economy by country of origin, the USA ranks consistently lower than both Europe and Japan, producing vehicles with the least fuel efficiency. European vehicles follow with intermediate fuel efficiency, while Japanese vehicles stand out as superior, consistently exhibiting higher fuel economy across all statistics. This pattern suggests a noteworthy disparity in fuel efficiency practices and priorities among these automotive regions. 

### Task 4

Create a visual showing how fuel economy has changed over time

```python
sns.set(style="whitegrid")  # Set Seaborn style
plt.figure(figsize=(12, 6))  # Set the figure size

# Seaborn lineplot
sns.lineplot(x='model_year', y='mpg', data=mpg.groupby('model_year')['mpg'].mean().reset_index(), marker='o', label='Average MPG')

# Matplotlib customization
plt.title('Average Fuel Economy Over Model Years')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend()

# Show the plot
plt.show()
```

Since the 1970s, there has been noticeable fluctuation in average miles per gallon (mpg) over the years, reflecting periods of both improvement and decline in vehicle fuel efficiency. Despite these fluctuations, an evident overall growth trend in fuel economy can be observed, and with advancing technology, it is anticipated that future vehicles will continue to exhibit a sustained increase in efficiency, aligning with the ongoing emphasis on technological advancements for improved environmental sustainability and fuel conservation.

### Task 5

Create two other visuals or tables related to fuel economy.

```python
# Plotting with Seaborn and Matplotlib
sns.set(style="whitegrid")  # Set Seaborn style
plt.figure(figsize=(10, 6))  # Set the figure size

# Seaborn histogram
sns.histplot(mpg['mpg'], bins=20, kde=True)

# Matplotlib customization
plt.title('Distribution of Fuel Economy')
plt.xlabel('Fuel Economy (MPG)')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

Examining the histogram, it becomes apparent that vehicles with the lowest and highest miles per gallon (MPG) are represented by the lowest counts, while the majority of vehicles fall within the range of 15 MPG to 35 MPG. This concentration in the middle range suggests that a substantial proportion of vehicles in the dataset exhibit moderate fuel efficiency, with fewer extremes in either very low or very high MPG categories.






### Task 6

```python
# Plotting with Seaborn and Matplotlib
sns.set(style="whitegrid")  # Set Seaborn style
plt.figure(figsize=(12, 8))  # Set the figure size

# Seaborn Violin Plot
sns.violinplot(x='origin', y='mpg', data=mpg, palette='Set2', inner='quartile')

# Matplotlib customization
plt.title('Violin Plot of Fuel Economy by Country of Origin')
plt.xlabel('Country of Origin')
plt.ylabel('Fuel Economy (MPG)')

# Show the plot
plt.show()
```

Add comments to your notebook with your observations.

In the violin plot comparing fuel economy by country of origin, the distribution for the USA is centered around 15-20 mpg, indicating that a significant number of American cars fall within this range. Conversely, Japanese cars exhibit a distribution peak at 30-35 mpg, reflecting a tendency towards higher fuel efficiency. European cars show a concentration in the 25-30 mpg range, suggesting a moderate level of fuel efficiency compared to both the USA and Japan.
